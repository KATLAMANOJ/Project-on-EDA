```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re
```

```python
plots = []
prices = []
Construction = []
totalarea = []
area = []
location = []
cityname=[]
builder = []
status = []

for x in range(3,25):
    url = "https://www.makaan.com/listings?beds=1,2,3,3plus&propertyType=apartment,builder-floor,villa,residential-plot,independent-house,studio-apartment&budget=0,&furnished=unfurnished,semifurnished,furnished&listingType=buy&pageType=CITY_URLS&cityName=Hyderabad&cityId=12&templateId=MAKAAN_CITY_LISTING_BUY&page={}".format(x)
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    for x in soup.find_all('div',class_ = 'title-line'):
        plots.append(x.text)
    
    for x in soup.find_all("td",class_="price"):
        prices.append(x.text)
    
    for x in soup.find_all("td",class_="val"):
        Construction.append(x.text)
    
    for x in soup.find_all("td",class_="size"):
        totalarea.append(x.text)
    
    for x in soup.find_all("td",class_="lbl rate"):
        area.append(x.text)
    
    
    for x in soup.find_all('span',class_="cityName"):
        cityname.append(x.text)
    
    
    for x in soup.find_all("span",itemprop="addressLocality"):
       location.append(x.text)
    
    
    
    for x in soup.find_all("div",class_="seller-info"):
        builder.append(x.text)
        m = []
        for val in builder:
            bui = re.compile("BUILDER0",re.I)    
            m.append(bui.sub("",str(val)))
    
for x in range(3,25):
    url = "https://www.makaan.com/bangalore-residential-property/buy-property-in-bangalore-city?beds=1,2,3,3plus&propertyType=apartment,builder-floor,villa,residential-plot,independent-house,studio-apartment&budget=0,&furnished=unfurnished,semifurnished,furnished&page={}&_=1673964857131".format(x)
    page = requests.get(url)
    soup = BeautifulSoup(page.text)
    for x in soup.find_all('div',class_ = 'title-line'):
        plots.append(x.text)
    
    for x in soup.find_all("td",class_="price"):
        prices.append(x.text)
    
    for x in soup.find_all("td",class_="val"):
        Construction.append(x.text)
    
    for x in soup.find_all("td",class_="size"):
        totalarea.append(x.text)
    
    for x in soup.find_all("td",class_="lbl rate"):
        area.append(x.text)
    
    for x in soup.find_all('span',class_="cityName"):
        cityname.append(x.text)
    
    
    for x in soup.find_all("span",itemprop="addressLocality"):
       location.append(x.text)
    
    for x in soup.find_all("div",class_="seller-info"):
        builder.append(x.text)
        m = []
        for val in builder:
            bui = re.compile("BUILDER0",re.I)    
            m.append(bui.sub("",str(val)))
     
    
```

```python
df3 = pd.DataFrame({"Plot_Specification":plots,"Prices":prices,"Construction Status":Construction,"Total Area":totalarea,"Price per Sq ft":area,"Location":location,"city":cityname,"Builder":m})
```

# converting prices into lakhs

```python
rate=[]
for i in df3['Prices']:
    if i.split()[-1]=='Cr':
        rate.append(round(float(i.split()[0])*100,2))
    else:
        rate.append(float(i.split()[0]))
```

```python
df3['price(in_lakhs)']=rate
df3.drop('Prices',axis=1,inplace=True)
```
